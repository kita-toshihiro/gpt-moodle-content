```python
# マイドライブに動画をアップロードしてください。それが「なにかの動画.mp4」という名前だとします。
# 「なにかの動画.mp4」から音声データのみを mp3 として取り出す
!ffmpeg -i /content/drive/MyDrive/なにかの動画.mp4  -f mp3 -ab 256k /content/drive/MyDrive/audio-ws1029.mp3
# whisperでの文字起こし（20分間の音声の場合、10分間程度で出来た。）
!whisper /content/drive/MyDrive/audio-ws1029.mp3 --language Japanese --device cuda --output_dir /content/drive/MyDrive/ --model large
```
